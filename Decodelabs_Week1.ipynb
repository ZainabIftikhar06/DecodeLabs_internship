```python
# Project 1: The Rule-Based AI Chatbot (The Logic Skeleton)

def main():
    #Defining KB using a Dictionary
    responses = {
        "hello": "Hi there! Welcome. How can I help you today?",
        "hi": "Hello! What can I do for you today?",
        "help": "I can assist you with basic queries. Try asking about 'project 1', 'internship', or 'ai'.",
        "project 1": "Project 1 is the Rule-Based Chatbot phase, focusing on control flow and logic.",
        "internship": "This internship batch is powered by DecodeLabs to build real-world AI portfolios.",
        "ai": "Artificial Intelligence starts with deterministic logic before moving to probabilistic models."
    }
    
    print("==================================================")
    print("         RULE-BASED AI CHATBOT ONLINE      ")
    print("      (Type 'exit' or 'bye' to terminate)         ")
    print("==================================================")
    
    #Heartbeat (Continuous Infinite Loop)
    while True:
        raw_input = input("\nYou: ")
        clean_input = raw_input.lower().replace(" ", "")
        
        #Exit Strategy (Kill Command)
        if clean_input in ['exit', 'bye']:
            print("Chatbot: Thank you for practicing! Deactivating logic engine. Goodbye!")
            break  
            
        # If the input isn't in our dictionary, print the deafult message (Using the atomic .get() method)
        reply = responses.get(clean_input, "Chatbot: I am sorry, I do not understand that command. Try asking 'help'.")
        
        print(reply)

if __name__ == "__main__":
    main()
```
